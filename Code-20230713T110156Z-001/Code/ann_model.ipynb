```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
features_list=['head_x_absEnergy', 'head_y_absEnergy', 'torso_x_absEnergy', 'torso_y_absEnergy', 'legs_x_absEnergy', 'legs_y_absEnergy','head_x_mean', 'head_y_mean', 'torso_x_mean', 'torso_y_mean', 'legs_x_mean', 'legs_y_mean','head_x_sd', 'head_y_sd', 'torso_x_sd', 'torso_y_sd', 'legs_x_sd', 'legs_y_sd','head_x_skewedness', 'head_y_skewedness', 'torso_x_skewedness', 'torso_y_skewedness', 'legs_x_skewedness', 'legs_y_skewedness', ]
print(features_list)
```

```python
# Read the dataset
X = pd.read_excel("/content/drive/MyDrive/Project Gait Analysis/Code /Extracted_features.xlsx")
X.head(84)
```

```python
# Define the labels
labels = ['wl', 'wl', 'wl', 'wl', 'wyc', 'wyc', 'wyc', 'wyc', 'fyc', 'fyc', 'fyc', 'fyc', 'hy', 'hy', 'hy', 'hy', 'ljg', 'ljg', 'ljg', 'ljg', 'lqf', 'lqf', 'lqf', 'lqf', 'lsl', 'lsl', 'lsl', 'lsl', 'ml', 'ml', 'ml', 'ml', 'nhz', 'nhz', 'nhz', 'nhz', 'rj', 'rj', 'rj', 'rj', 'syj', 'syj', 'syj', 'syj', 'wl', 'wl', 'wl', 'wl', 'wq', 'wq', 'wq', 'wq', 'xch', 'xch', 'xch', 'xch', 'xxj', 'xxj', 'xxj', 'xxj', 'yjf', 'yjf', 'yjf', 'yjf', 'zc', 'zc', 'zc', 'zc', 'zdx', 'zdx', 'zdx', 'zdx', 'zjg', 'zjg', 'zjg', 'zjg', 'zl', 'zl', 'zl', 'zl', 'zyf', 'zyf', 'zyf', 'zyf']
Y = pd.DataFrame(labels)
```

```python

# Encode labels as integers
label_encoder = LabelEncoder()
Y_encoded = label_encoder.fit_transform(Y.values.ravel())
```

```python
# Convert labels to one-hot vectors
num_classes = len(label_encoder.classes_)
Y_one_hot = to_categorical(Y_encoded, num_classes=num_classes)
```

```python
# Split dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_one_hot, test_size=0.2, random_state=0)
```

```python
# Perform feature scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Define the ANN model
model = Sequential()
model.add(Dense(64, input_shape=(24,), activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))
```

```python
# Compile the ANN
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=['accuracy'])
```

```python
# Fit the ANN
model.fit(X_train, Y_train, batch_size=32, epochs=100)
```

```python
loss, accuracy = model.evaluate(X_test, Y_test)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```
